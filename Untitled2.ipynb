```
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
def f1(x):
  return x*0-2
x = np.linspace(-6,-4,100)
y = f1(x)
plt.figure(figsize=(5,5))
plt.plot(x,y,'-r', linewidth=3)
plt.grid(True)

```

```
def f2(x):
 return (0.25*x)
x = np.linspace(-4,0,100)
y = f2(x)
plt.figure(figsize=(5,5))
plt.plot(x,y,'-r', linewidth=3)
plt.grid(True)
```

```
def f3(x):
  return 1-np.sqrt(1-x*x)
x = np.linspace(0,1,100)
y = f3(x)
plt.figure(figsize=(5,5))
plt.plot(x,y,'-r', linewidth=3)
plt.grid(True)
```

```
def f4(x):
  return 3*x-2
x = np.linspace(1,2,100)
y = f4(x)
plt.figure(figsize=(5,5))
plt.plot(x,y,'-r', linewidth=3)
plt.grid(True)
```

```
def f5(x):
   return -0.5*x+5
x = np.linspace(2,12,100,)
y = f5(x)
plt.figure(figsize=(5,5))
plt.plot(x,y,'-r', linewidth=3)
plt.grid(True)
```

```
def foo(x):
    if (x >= -6 and x < -4 ):
        return f1(x)
    elif (x >= -4 and x < 0 ):
        return f2(x)
    elif (x >= 0 and x < 1 ):
        return f3(x)
    elif (x >= 1 and x < 2 ):
        return f4(x)
    elif (x >= 2 and x < 12 ):
        return f5(x)
x = np.linspace(-6,12,1000)
y = [foo(i) for i in x]
plt.figure(figsize=(12,4))
plt.plot(x,y,'-r', linewidth=3)
plt.grid(True)
```
